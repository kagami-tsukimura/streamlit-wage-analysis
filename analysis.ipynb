#### データの可視化

```python
df_ts_mean = df_jp_ind[df_jp_ind["年齢"] == "年齢計"]
df_ts_mean
```

#### 年齢階級別の一人あたり賃金を箱ひげ図表示

```python
print(df_pref_category.shape)
df_pref_category.head()
```

```python
ages = df_pref_category["年齢"].unique()
ages
```

```python
wages = []
for age in ages:
    wage_temp = df_pref_category[df_pref_category["年齢"] == age]["一人当たり賃金（万円）"].values.tolist()
    wages.append(wage_temp)
```

```python
len(wages)
```

```python
len(wages[0])
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title("年齢階級ごとの一人あたり賃金")
ax.set_xticklabels(ages)
ax.boxplot(wages)
plt.show()
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title("年齢階級ごとの一人あたり賃金")
ax.set_xticklabels(ages)
ax.boxplot(wages, showfliers=False)
plt.show()
```

#### 産業別の一人あたり賃金を箱ひげ図表示

```python
inds = df_pref_category["産業大分類名"].unique()
inds
```

```python
wages = []
for ind in inds:
    wage_temp = df_pref_category[df_pref_category["産業大分類名"] == ind]["一人当たり賃金（万円）"].values.tolist()
    wages.append(wage_temp)
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title("産業別の一人あたり賃金")
ax.set_xticklabels(inds, rotation=80)
ax.boxplot(wages, showfliers=False)
plt.show()
```
